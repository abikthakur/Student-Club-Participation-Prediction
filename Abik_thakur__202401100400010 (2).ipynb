```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = pd.DataFrame({
    'interest_level': [4, 6, 8, 6, 9],
    'free_hours_per_week': [17, 12, 19, 19, 17],
    'club_participation': ['no', 'no', 'no', 'yes', 'no']
})

# Plotting a simple bar graph of interest level vs free hours
plt.figure(figsize=(8, 5))
plt.bar(data.index, data['interest_level'], color='skyblue', label='Interest Level')
plt.plot(data.index, data['free_hours_per_week'], color='green', marker='o', label='Free Hours/Week')

# Labels and legend
plt.title('Interest Level and Free Hours (per student)')
plt.xlabel('Student Index')
plt.ylabel('Value')
plt.xticks(data.index)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
